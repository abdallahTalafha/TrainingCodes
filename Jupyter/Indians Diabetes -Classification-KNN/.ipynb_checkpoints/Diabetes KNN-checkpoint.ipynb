#fill mising data
 - Gluocose
 - BloodPressure

 - Insulin

```python
x = data[data['Insulin']!=0]
x['Insulin'].median()
```

```python
def fill_missing_data(col,value):
    t=data[col]
    for i in t:
        if i ==0:
            i = value
    return col
```

```python
def calculate_Value(col_name,method_name):
    x=data[col_name]
    x=x[x!=0]
    if method_name == 1:
        return x.mean()
    elif method_name == 2:
        return x.mode()
    return x.median()
```

```python
data['Glucose']= fill_missing_data('Glucose',calculate_Value('Glucose',1))
```

```python
grid = sns.FacetGrid(data, col='Outcome',height=5,aspect=1)
grid.map(plt.hist,'Age',alpha=.5, bins=30)
grid.add_legend()
p = data.hist(figsize = (20,20)
```
